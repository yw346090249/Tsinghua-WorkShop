#### Step 2 - Rich Output
In addition to text output, Jupyter notebooks can produce rich graphical content.

Run the code in the following cell.  It will display a slider bounded from 2 to 20.  Following the slider will be a square filled with randomly colored blocks.  Manipulating the slider will interactively change the number of blocks per edge within the square.

```pyspark
%matplotlib inline
import numpy as np;
import matplotlib.pyplot as plt;
from ipywidgets import interact;
@interact(x=(2,20))
def square(x):
    plt.imshow(np.random.rand(x,x), interpolation='none');
```

#### Step 3 - Magic Commands
Jupyter provides several *magic commands* that provide additional functionality above within a given line or cell.  *Line-magics* operate on a single line only and are preceded by a single `%`.  *Cell-magics* operate on the entire contents of the cell, must be the first entry in the cell, and are preceded by a double `%%`.

In the previous step you saw the `%matplotlib inline` line-magic used to direct the output of the matplotlib drawing library to be inline within the notebook.

Run the code in the following cell.  It uses the `%%sql` magic command to run a SQL query against the *HiveSampleTable* sample Hive database table that stored in the cluster's default storage.  Notice how you can interact with the results of a SQL query.  You will learn more about this kind of data investigation and interaction in the next Exercise.

```pyspark
%%sql
SELECT DevicePlatform, Count(*) as total 
FROM HiveSampleTable 
WHERE Country = 'United States' 
GROUP BY DevicePlatform 
ORDER BY Total DESC
```

You can list the available magic commands by using the `%lsmagic` command.  To obtain information about a specific magic command, append a question mark (?) after the command.

Run the code in the following cell.  It will open the help panel for the `%%sql` magic command.  When you are done, use the **x** icon in the panel's upper right corner to close it.

```pyspark
%%sql?
```
